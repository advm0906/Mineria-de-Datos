## Base de datos 1
- *Nombre:* Bitcoin Historical Dataset
- *Url:* *[Bitcoin](https://www.kaggle.com/prasoonkottarathil/btcinusd)*

Primero visualizaremos la base de datos natural

```python
#Importamos las bibliotecas que nos ayudaran
import pandas as pd
import numpy as np
import json
#Ahora abriremos el archivo para visualizarlo 
bitcoin=pd.read_csv("C:/Users/HP/Documents/Facultad/Mineria/PIA/BTCUSD_day.csv",encoding='utf8')
bitcoin
```

Para continuar usaremos una base de datos limpia, por lo que haremos limpieza de datos

```python
#se eliminan las columnas que no son de utilidad
del(bitcoin['Symbol'])
del(bitcoin['Volume USD'])
#no funciono con la funcion de .drop() y recurrimos a usar la de del()
#Renombramos las columnas a español para facilitar el entendimiento de nuestro equipo y lectores
bitcoin.rename(columns= {'Date': 'Fecha','Open':'Apertura','High':'Máximo','Low': 'Mínimo','Close': 'Cierre','Volume BTC': 'Transacciones_BTC'},inplace=True)
bitcoin
```

### *Información estadística básica*
#### Con la base de datos limpia ahora continuaremos con la información Estadistica

Jugaremos con cada una de las opciones que maneja Pandas para la estadística básica para algunas columnas.
- Iniciamos calculando la suma de las columnas de Apertura, Máximo, Mínimo, Cierre y Transacciones_BTC

```python
print("Suma de Apertura: ", bitcoin['Apertura'].sum())
print("Suma de Máximo: ", bitcoin['Máximo'].sum())
print("Suma de Mínimo: ", bitcoin['Mínimo'].sum())
print("Suma de Cierre: ", bitcoin['Cierre'].sum())
print("Suma de Transacciones_BTC: ", bitcoin['Transacciones_BTC'].sum())
```

- Promedio de las columnas de Apertura, Máximo, Mínimo, Cierre y Transacciones_BTC

```python
print("Promedio de Apertura: ", bitcoin['Apertura'].mean())
print("Promedio de Máximo: ", bitcoin['Máximo'].mean())
print("Promedio de Mínimo: ", bitcoin['Mínimo'].mean())
print("Promedio de Cierre: ", bitcoin['Cierre'].mean())
print("Promedio de Transacciones_BTC: ", bitcoin['Transacciones_BTC'].mean())
```

- Suma acumulada de la columna fila por fila de las columnas de Apertura y Cierre

```python
print("Suma acumulada de Apertura: ", bitcoin['Apertura'].cumsum())
print(" ")
print("Suma acumulada de Cierre: ", bitcoin['Cierre'].cumsum())
```

- Resumen estadístico de las columna de Transacciones_BTC

```python
print("Resumen estadístico de Transacciones_BTC: ", bitcoin['Transacciones_BTC'].describe())
```

- Cuantos elementos no nulos hay en la columna de Apertura y Cierre

```python
print("Elementos no nulos de Apertura: ", bitcoin['Apertura'].count())
print("Elementos no nulos de Cierre: ", bitcoin['Cierre'].count())
```

- Minimo y máximo de las columnas de Máximo y Mínimo

```python
print("Valor máximo de la columna Máximo: ", bitcoin['Máximo'].max())
print("Valor mínimo de la columna Máximo: ", bitcoin['Máximo'].min())
print("")
print("Valor máximo de la columna Mínimo: ", bitcoin['Mínimo'].max())
print("Valor mínimo de la columna Mínimo: ", bitcoin['Mínimo'].min())
```

- Ahora calcularemos la mediana, varianza, desviación estandar, valor de asimetría en los datos y la característica de forma de su distribución de frecuencias/probabilidad de cada una de las columnas

```python
print("           Apertura")
print("Media:               ", bitcoin['Apertura'].median())
print("Varianza:            ", bitcoin['Apertura'].var())
print("Desviación estandar: ", bitcoin['Apertura'].std())
print("Valor de asimetría:  ", bitcoin['Apertura'].skew())
print("Forma distribucion: ", bitcoin['Apertura'].kurt())
print("")

print("           Máximo")
print("Media:               ", bitcoin['Máximo'].median())
print("Varianza:            ", bitcoin['Máximo'].var())
print("Desviación estandar: ", bitcoin['Máximo'].std())
print("Valor de asimetría:  ", bitcoin['Máximo'].skew())
print("Forma distribucion: ", bitcoin['Máximo'].kurt())
print("")

print("           Mínimo")
print("Media:               ", bitcoin['Mínimo'].median())
print("Varianza:            ", bitcoin['Mínimo'].var())
print("Desviación estandar: ", bitcoin['Mínimo'].std())
print("Valor de asimetría:  ", bitcoin['Mínimo'].skew())
print("Forma distribucion: ", bitcoin['Mínimo'].kurt())
print("")

print("           Cierre")
print("Media:               ", bitcoin['Cierre'].median())
print("Varianza:            ", bitcoin['Cierre'].var())
print("Desviación estandar: ", bitcoin['Cierre'].std())
print("Valor de asimetría:  ", bitcoin['Cierre'].skew())
print("Forma distribucion: ", bitcoin['Cierre'].kurt())
print("")

print("       Transacciones_BTC")
print("Media:               ", bitcoin['Transacciones_BTC'].median())
print("Varianza:            ", bitcoin['Transacciones_BTC'].var())
print("Desviación estandar: ", bitcoin['Transacciones_BTC'].std())
print("Valor de asimetría:  ", bitcoin['Transacciones_BTC'].skew())
print("Forma distribucion: ", bitcoin['Transacciones_BTC'].kurt())
```

- Por último revisaremos la correlación y la covarianza entre los datos de las columnas de nuestra base de datos

```python
bitcoin.corr()
```

```python
bitcoin.cov()
```
