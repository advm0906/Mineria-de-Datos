### *?df.dropna()*

DataFrame.dropna( eje = 0 , cómo = 'cualquiera' , umbral = Ninguno , subconjunto = Ninguno , en el lugar = Falso )     
Elimina los valores perdidos


Parámetros : 

- eje {0 o 'índice', 1 o 'columnas'}: determina si se eliminan las filas o columnas que contienen valores faltantes. 
0, o 'índice': elimine las filas que contienen valores faltantes y 1, o 'columnas': elimine las columnas que contengan un valor faltante.


- cómo {'cualquiera', 'todos'}: determina si la fila o columna se elimina del DataFrame, cuando tenemos al menos un NA o todos los NA.
'any': si hay valores NA presentes, elimine esa fila o columna y 'all': si todos los valores son NA, elimine esa fila o columna.


- thresh int, opcional: Requiere tantos valores que no sean NA.


- subconjunto similar a una matriz, opcional: etiquetas a lo largo de otro eje a considerar, por ejemplo, si está eliminando filas, estas serían una lista de columnas para incluir.


- inplace bool: Si es Verdadero, realice la operación en el lugar y devuelva Ninguno


### *?df.astype*

DataFrame.astype( dtype , copy = True , errors = 'raise' )        
Transmite un objeto pandas a un dtype especificado dtype.

Parámetros:

- dtype: tipo de datos, o dict del nombre de la columna -> tipo de datos
Use un tipo numpy.dtype o Python para convertir el objeto pandas completo en el mismo tipo. Alternativamente, use {col: dtype,…}, donde col es una etiqueta de columna y dtype es un tipo numpy.dtype o Python para convertir una o más de las columnas del DataFrame en tipos específicos de columna.


- copy: bool
Devolver una copia cuando copy=True(tenga mucho cuidado con la configuración, copy=Falseya que los cambios en los valores pueden propagarse a otros objetos pandas).


- errores {'subir', 'ignorar'}: Controle la generación de excepciones sobre datos no válidos para el tipo de datos proporcionado.
raise : permitir que se generen excepciones
ignore: suprime las excepciones. En caso de error, devuelva el objeto original.
